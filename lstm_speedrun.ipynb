```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import TensorDataset, DataLoader

VOCAB_SIZE = 501
EMBEDDING_DIM = 100
SEQUENCE_LENGTH = 44
LSTM_UNITS = 150
OUTPUT_CLASSES = 501

class LSTMMultiLayer(nn.Module):
    def __init__(self):
        super().__init__()

        # 1. Embedding Layer: (B, 44) -> (B, 44, 100)
        self.embedding = nn.Embedding(VOCAB_SIZE, EMBEDDING_DIM)

        # 2. First LSTM Layer: (B, 44, 100) -> (B, 44, 150)
        # return_sequences=True equivalent is the default 'output' of nn.LSTM
        self.lstm1 = nn.LSTM(
            input_size=EMBEDDING_DIM,
            hidden_size=LSTM_UNITS,
            batch_first=True,
        )

        # 3. Second LSTM Layer: (B, 44, 150) -> Final output (1, B, 150)
        self.lstm2 = nn.LSTM(
            input_size=LSTM_UNITS,
            hidden_size=LSTM_UNITS,
            batch_first=True
        )

        # 4. Dense/Linear Layer: (B, 150) -> (B, 501)
        # Input size is the hidden size of the last LSTM's final state (150)
        self.dense = nn.Linear(LSTM_UNITS, OUTPUT_CLASSES)

        # RECOMMENDATION: We REMOVE the nn.Softmax layer here
        # because nn.CrossEntropyLoss handles the LogSoftmax internally
        # for better numerical stability.

    def forward(self, x):
        # x shape: (Batch Size, Sequence Length) e.g., (32, 44)

        # 1. Embedding
        x = self.embedding(x)
        # x shape: (32, 44, 100)

        # 2. First LSTM (passes full sequence to the next LSTM)
        output1, _ = self.lstm1(x)
        # output1 shape: (32, 44, 150)

        # 3. Second LSTM
        # We only care about the final hidden state (h_n) for the classification task.
        # h_n shape: (num_layers=1, Batch Size, Hidden Size) -> (1, 32, 150)
        _, (h_n, _) = self.lstm2(output1)

        # Select the final hidden state and remove the layer dimension
        # x shape: (32, 150)
        x = h_n.squeeze(0)

        # 4. Dense/Linear
        # x shape: (32, 501) - These are the raw logits
        x = self.dense(x)

        return x

# --- 3. Prepare Data ---
BATCH_SIZE = 32
N_SAMPLES = 861

X_data = torch.randint(0, VOCAB_SIZE, (N_SAMPLES, SEQUENCE_LENGTH), dtype=torch.long)

Y_data = torch.randint(0, OUTPUT_CLASSES, (N_SAMPLES,), dtype=torch.long)

dataset = TensorDataset(X_data, Y_data)
dataloader = DataLoader(dataset, batch_size=BATCH_SIZE, shuffle=True)

model = LSTMMultiLayer()

optimizer = optim.Adam(model.parameters())

# nn.CrossEntropyLoss is used because the model outputs raw logits (no softmax).
criterion = nn.CrossEntropyLoss()

NUM_EPOCHS = 50

print(f"Starting training for {NUM_EPOCHS} epochs...")
for epoch in range(NUM_EPOCHS):
    model.train()
    total_loss = 0
    total_correct = 0
    total_samples = 0

    for batch_idx, (inputs, labels) in enumerate(dataloader):

        optimizer.zero_grad()

        # Forward pass
        outputs = model(inputs)

        # Calculate loss
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        total_loss += loss.item() * inputs.size(0)

        # Calculate accuracy
        _, predicted = torch.max(outputs.data, 1)
        total_samples += labels.size(0)
        total_correct += (predicted == labels).sum().item()

    avg_loss = total_loss / total_samples
    avg_accuracy = total_correct / total_samples

    print(f'Epoch [{epoch+1}/{NUM_EPOCHS}] | Loss: {avg_loss:.4f} | Accuracy: {avg_accuracy:.4f}')

print("Training finished.")
```

```python
inv_vocab = {v: k for k, v in vocab.items()}

def predict_next_word(model, text, vocab, inv_vocab, max_len=44):
    model.eval()
    words = text.split()

    for i in range(1):  # generate x words
        # convert to indices
        tokenized = [vocab.get(w, 0) for w in words]  # 0 for unknown words
        # pad to max_len
        padded = [0] * (max_len - len(tokenized)) + tokenized
        padded_tensor = torch.tensor([padded], dtype=torch.long)

        # predict
        with torch.no_grad():
            output = model(padded_tensor)
            # output shape: (batch_size, vocab_size)
            pos = torch.argmax(output, dim=1).item()

        # map predicted index to word
        next_word = inv_vocab.get(pos, "")
        words.append(next_word)
        text = " ".join(words)
        # print(text)
        time.sleep(1)

    return text
```
